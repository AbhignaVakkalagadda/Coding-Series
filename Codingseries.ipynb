```python
def fact(n):
  if n==0 or n==1:
    return 1
  return n*fact(n-1)
print(fact(5))
```

```python
def fact(n,t=1):
  if n==0 or n==1:
    return t
  return fact(n-1,n*t)
print(fact(5))
```

```python
''' code for math operations using base class , with 1st and 2nd derived class
respectively .
create an object where derived will be executed and perform multiple operaions '''
class numbers:
    def __init__(self, a,b):
        self.a=a
        self.b=b
class basicops(numbers):
    def add(self):
        return self.a+self.b
    def sub(self):
        return self.a-self.b
class advops(basicops):
    def mul(self):
        return self.a*self.b
    def div(self):
        return self.a/self.b
x= int(input("enter the value of x:"))
y= int(input("enter the value of y:"))
calc = advops(x,y)
print("Addition:", calc.add())
print("Subtract:", calc.sub())
print("Product:", calc.mul())
print("Divide:", calc.div())
```

```python
''' code for student grading system , where student base class has to
be derived by marks and marks with grades in a multi-level inheritance .
caliculate the grade to be alloted to the student with marks A,B,C,D with >=90,
>=75, >=60 or less than 60 respectively .
student: Henry
marks : [85, 90, 78, 92]
caliculate and display marks, avg, grade of Henry '''
class student:
    def __init__(self,name):
        self.name= name
class marks(student):
    def __init__(self,name,marks):
        super().__init__(name)
        self.marks= marks
    def total_marks(self):
        return sum(self.marks)
class grade(marks):
    def average(self):
        return self.total_marks()/len(self.marks)
    def cal_grade(self):
        avg= self.average()
        if avg>=90:
            return 'A'
        elif avg>=75:
            return 'B'
        elif avg>=60:
            return 'C'
        else:
            return 'D'
name= input("enter the student name")
subjects=int(input("enter the number of subjects"))
marks= []
for i in range(subjects):
    mark= float(input(f"enter the marks for subject{i+1}:"))
    marks.append(mark)
student=grade(name, marks)
print("Student Name:",student.name)
print("Total marks:",student.total_marks())
print("Average:",student.average())
print("Student Grade:",student.cal_grade())
```

```python
''' Hierarchical geometry operations of different shapes area values of all
circle , rectangle , triangle derived classes '''
import math
class shape:
    def __init__(self, d1, d2=0):
        self.d1=d1
        self.d2=d2
class rectangle(shape):
    def area(self):
        return self.d1 * self.d2
class circle(shape):
    def area(self):
        return math.pi * self.d1 *self.d1
class triangle(shape):
    def area(self):
        return 0.5 *self.d1 * self.d2
rect= rectangle(10,5)
cir= circle(7)
tri= triangle(8,6)
print("Area of rectangle:", rect.area())
print("Area of circle:", cir.area())
print("Area of triangle:", tri.area())
```

```python
# abstraction of the base class for evaluating the area of sq, rect, tri
from abc import ABC, abstractmethod
from math import pi
class shape(ABC):
    def __init__(self,name):
        self.name=name
    @abstractmethod
    def area(self):
        pass
class rectangle(shape):
    def __init__(self, l, b):
        super().__init__("rectangle")
        self.l=l
        self.b=b
    def area(self):
        return self.l * self.b
class circle(shape):
    def __init__(self, r):
        super().__init__("circle")
        self.r=r
    def area(self):
        return pi *self.r * self.r
class triangle(shape):
    def __init__(self, b, h):
        super().__init__("triangle")
        self.h=h
        self.b=b
    def area(self):
        return 0.5 * self.b * self.h
r= rectangle(11,2)
c=circle(6)
t= triangle(4,6)
print("Area of rectangle:",r.area())
print("Area of circle:",round(c.area(), 2))
print("Area of triangle:",t.area())
```

```python
def binary_search(arr,key):
  low=0
  high=len(arr)-1
  while low<=high:
    mid=(low+high)//2
    if arr[mid]==key:
      return mid
    elif arr[mid]<key:
      low=mid+1
    else:
      high=mid-1
  return -1
size=int(input("Enter the size of the array:"))
arr=[]
print("Enter the elements:")
for i in range(size):
  num=int(input(f"Element {i+1}:"))
  arr.append(num)
key=int(input("Enter the element to search"))
result=binary_search(arr,key)
if result !=1:
  print(f"{key} is present in the array and found at {result}")
else:
  print(f"{key} is not present in the array")
```

jump search

1. size  n sorted  sorted

2. create a block key
3. search  operation will be performed
arr[m]<var/key<arr(k+1)[m]
4.compare  each jump linearly
5.time complexity O(^N)
6.space complexity O(1)

```python
import math
def jump_search(arr,target):
  if not arr:
    return -1

  n=len(arr)
  step=int(math.sqrt(n))
  prev=0
  while prev <n and arr[prev]<target:
    prev +=step
  for i in range (max(0,prev-step),min(n,prev)):
    if arr[i]==target:
      return i
  return -1
arr=[1,3,5,7,8,9,11]
target=7
result=jump_search(arr,target)
print(f"Eelement {target} found at index {result}")
```

Exponential search

1. sorted numbers searching

2. check the array !=0
3. check the first element
4. find the range using the expo while for boundary
identify within boundary
5. range perform binary search
6. return reult

```python
def bsearch_range(arr,target,left,right):
  while left<=right:
    mid=(left+right)//2
    if arr[mid]==target:
      return mid
    elif arr[mid]<target:
      left=mid+1
    else:
      right=mid-1
  return -1
def expo_search(arr,target):
  if not arr:
    return -1
  if arr[0]==target:
    return 0
  n=len(arr)
  i=1
  while i<n and arr[i]<=target:
    i *=2
  return bsearch_range(arr,target,i//2,min(i,n-1))
arr=[2,4,6,8,10,12,14]
target=10
result=expo_search(arr,target)
print(f"Element {target} found at index {result}")
```

Fibonacci search
1. find smallest fibo number


```python
def fibsearch(arr, target):
    if not arr:
        return -1
    n = len(arr)
    fib2 = 0
    fib1 = 1
    fib = fib1 + fib2
    while fib < n:
        fib2 = fib1
        fib1 = fib
        fib = fib1 + fib2
    offset = -1
    while fib > 1:
        i = min(offset + fib2, n - 1)
        if arr[i] < target:
            fib = fib1
            fib1 = fib - fib2
            fib2 = fib - fib1
            offset = i
        elif arr[i] > target:
            fib = fib2
            fib1 = fib1 - fib2
            fib2 = fib - fib1
        else:
            return i
    if fib1 == 1 and offset + 1 < n and arr[offset + 1] == target:
        return offset + 1
    return -1

arr = [2, 4, 6, 8, 10, 12]
target = 10
result = fibsearch(arr, target)
print(f"Element {target} found at index {result}")
```

```python
def fib_search(arr, target):
    a, b = 0, 1
    while b < len(arr):
        a, b = b, a + b
    offset = -1
    while b > 1:
        i = min(offset + a, len(arr) - 1)
        if arr[i] < target:
            b, a = a, b - a
            offset = i
        elif arr[i] > target:
            b, a = a, b - a
        else:
            return i
    if a and offset + 1 < len(arr) and arr[offset + 1] == target:
        return offset + 1
    return -1

# Test
arr = [2, 4, 6, 8, 10, 12]
target = 10
print(f"Element {target} found at index {fib_search(arr, target)}")
```
