```python
def info(**kwargs):
  for key,value in kwargs.items():
    print(f"{key}:{value}")
info(name="Abhi",age=20,city="vijayawada")
```

**RECURSION**

**DIRECT: Recurisive factorial**

```python
def fact(n):
  if n==0 or n==1:
    return 1
  return n*fact(n-1)
print(fact(5))
```

**INDIRECT: Even or Odd**

```python
def e(n):
  if n==0:
    return True
  return o(n-1)
def o(n):
  if n==0:
    return False
  return e(n-1)
print(e(2))
print(o(3))
```

**TAIL :FACT**

```python
def fact(n,t=1):
  if n==0 or n==1:
    return t
  return fact(n-1,n*t)
print(fact(5))
```

**LINEAR**

```python
def total(n):
  if n<0:
    return 0
  return n+total(n-1)
print(total(10))
```

**NESTED** **SUPER FACTORIAL**

```python
def sfact(n):
  if n<=0:
    return 1
  return fact(n)*sfact(n-1)
def fact(n):
  if n<=1:
    return 1
  return n*fact(n-1)
print(sfact(4))
print(sfact(3))
```

**POWER TOWER**

```python
def pt(a,n):
  if n==1:
    return a
  return a**pt(a,n-1)
print(pt(2,4))
print(pt(3,2))
```

**Arithmetic opertaion by using a class obj instance** **Direct method**

```python
class calc:
  def add(self,a,b):
    return a+b
  def mul(self,a,b):
    return a*b
c=calc()
print(c.add(5,6))
print(c.mul(5,4))

```

**Asscessing methodology**

```python
class calc:
  def set_values(self,a,b):
    self.a=a
    self.b=b
  def add(self):
    return self.a+self.b
  def mul(self):
    return self.a*self.b
c=calc()
c.set_values(4,5)
print(c.add())
print(c.mul())
```

**Code for calculating the area of a square by creating a class square & defining the area and perimeter functions and return the both the values to the objects**

```python
class square:
  def area(self,s):
    return s*s
  def perimeter(self,s):
    return 4*s
S=square()
print(S.area(5))
print(S.perimeter(5))
```

```python
class square:
  def set_values(self,side):
    self.side=side
  def area(self):
    return self.side**2
  def perimeter(self):
    return 4*self.side
S=square()
S.set_values(5)
print(S.area())
print(S.perimeter())
```

```python
# innheritance multiple
# create a super class from a constructor and call the base class
class base1(object):
    def __init__(self):
        super(base1, self).__init__()
        print("Base class-1")
class base2(object):
    def __init__(self):
        super(base2, self).__init__()
        print("Base class-2")
class base3(object):
    def __init__(self):
        super(base3, self).__init__()
        print("Base class-3")
class derived(base1, base3, base2):
    def __init__(self):
        super(derived, self).__init__()
        print("Derived class")
d= derived()
```

```python
class addition:
    def add(self, a,b):
        return a+b
class subtract:
    def sub(self, a,b):
        return a-b
class calc(addition, subtract):
    def product(self, a, b):
        return a*b
c= calc()
n1= int(input("Enter the value of a :"))
n2= int(input("Enter the value of b :"))
print("Addition:", c.add(n1,n2))
print("Difference:", c.sub(n1,n2))
print("Multiplication:", c.product(n1,n2))
```
