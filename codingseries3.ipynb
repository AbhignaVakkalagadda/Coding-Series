```python
import pygame
import random
import sys
pygame.init()
height= 500
width = 500
block_size=15
pygame.display.set_caption('🐍Snake Game')
BLACK=(0,0,0)
WHITE=(255,255,255)
RED=(255,0,0)
GREEN=(0,255,0)
BLUE=(0,0,255)
screen = pygame.display.set_mode((width, height))
clock=pygame.time.Clock()
font=pygame.font.SysFont(None,36)
def draw_snake(snake_list):
  for block in snake_list:
    pygame.draw.rect(screen,GREEN,(block[0],block[1],block_size,block_size))
def draw_food(pos):
  pygame.draw.rect(screen,RED,(pos[0],pos[1],block_size,block_size))
def message(msg,color):
  text = font.render(msg,True,color)
  screen.blit(text,[width//4,height//2])
def game_loop():
  snake=[(100,100)]
  direction="RIGHT"
  change_to=direction
  food=[random.randrange(0,width//block_size)*block_size,random.randrange(0,height//block_size)*block_size]
  score=0
  while True:
    for event in pygame.event.get():
      if event.type==pygame.QUIT:
        pygame.quit()
        sys.exit()
      elif event.type==pygame.KEYDOWN:
        if event.key==pygame.K_UP and direction!="DOWN":
          change_to="UP"
        elif event.key==pygame.K_DOWN and direction!="UP":
          change_to="DOWN"
        elif event.key==pygame.K_LEFT and direction!="RIGHT":
          change_to="LEFT"
        elif event.key==pygame.K_RIGHT and direction!="LEFT":
          change_to="RIGHT"
    direction=change_to
    x,y=snake[0]
    if direction=="UP":
      y-=block_size
    elif direction=="DOWN":
      y+=block_size
    elif direction=="LEFT":
      x-=block_size
    elif direction=="RIGHT":
      x+=block_size
    head=(x,y)
    if x<0 or x>=width or y<0 or y>=height or head in snake:
      screen.fill(BLACK)
      message("Game Over!! Score:"+str(score),RED)
      pygame.display.update()
      pygame.time.wait(2000)
      return
    snake.insert(0,head)
    if head == tuple(food):
      score +=1
      food =[random.randrange(0,width//block_size)*block_size,random.randrange(0,height//block_size)*block_size]
    else:
      snake.pop()
    screen.fill(BLACK)
    draw_snake(snake)
    draw_food(food)
    score_text=font.render("Score:"+str(score),True,WHITE)
    screen.blit(score_text,[10,10])
    pygame.display.update()
    clock.tick(10)

game_loop()
```
